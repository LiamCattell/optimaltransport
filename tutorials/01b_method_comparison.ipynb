## Common Attributes and Functions

All of the continuous optimal transport methods the *optimaltransport* package are derived from a single base class `BaseTransform`. This means that some useful attributes and functions are common to all of the optimal transport algorithms. For example, all of the optimal transport classes have the following two attributes:

* `transport_map_`: The transport map $f$ between two signals $I_0$ and $I_1$. i.e. $I_0 = \det(D_f) I_1 \circ f$, where $D_f$ is the Jacobian of $f$.
* `displacements_`: The pixel displacements $u$, where $u(x) = x - f(x)$.

Similarly, all of the optimal transport classes include the following functions:

* `forward(sig0, sig1):` Computes the forward LOT transform ($\hat{I}_1 = u\sqrt{I_0}$) between two signals.
* `inverse()`: Computes the inverse LOT transform and returns the reconstructed signal $I_1$
* `apply_forward_map(transport_map, sig1)`: Apply a transport map $f$ to a signal $I_1$. Returns $I_0 = \det(D_f) I_1 \circ f$.
* `apply_inverse_map(transport_map, sig0)`: Compute the inverse of a transport map $f$, and apply it to reference signal $I_0$. Returns $I_1 = \det(D_f)^{-1} I_0 \circ f^{-1}$.

By sharing attributes and functions, you can switch between optimal transport methods with minimal hassle. Some methods have additional attributes and functions, and more information can be found in the documentation.

## Optimal Transport Algorithms

In this tutorial we will compare two-dimensional continuous optimal transport methods. The term "continuous" indicates that the methods are derived from the Monge formulation of the optimal transport problem, rather than Kantarovich's discrete formulation:

$$
M(f) = \int_X \|x-f(x)\|^2 I_0(x) dx \quad s.t. \; I_0(x) = \det(D_f(x)) I_1(f(x))
$$

Before we compare the results of the transforms, let's introduce each of the methods.

### `RadonCDT`

This is a 2D extension to the cumulative distribution transform (CDT) [1], and was covered in more detail in Tutorial 1a. Briefly, the Radon-CDT algorithm transforms an image into a set of 1D projections using the Radon transform, and then computes the 1D CDT of each projection [2].

### `CLOT`

The continuous linear optimal transport (CLOT) algorithm is a two-step approach to computing the optimal transport map between two images [3]. Firstly, an initial mapping is computed using the technique described in [4]. The curl is then removed from the mapping by minimizing the Monge optimal transport objective function via gradient descent.


### `VOT2D`

The variational optimal transport (VOT) method modifies the Monge objective function to help guide the solution towards a curl-free mapping [5]. The optimization problem is relaxed further to incorporate the mass-preserving constraint, resulting in the following objective function:

$$
\psi(f) = \frac{1}{2}\int_X \|x-f(x)\|^2 I_0(x)dx + \frac{\alpha}{2} \int_X \|\nabla \times f(x)\|^2 + \frac{1}{2} \int_X (\det(D_f(x)) I_1(f(x)) - I_0(x))^2 dx
$$

where $\alpha$ is a regularization parameter and $\nabla \times (\cdot)$ is the curl operator. The VOT objective function is non-convex, so the algorithm's authors use a variational optimization technique to solve it [5].


### `MultiVOT2D`

The VOT method described above does not theoretically guarantee that the global mimimum of the objective function will be found. As a result, a multi-scale technique is employed in [5] to guide the method towards the globally optimum solution. The `VOT` class computes the optimal transport map at a single scale, whereas `MultiVOT2D` uses the multi-scale approach.

### `SPOT2D`

The single-potential optimal transport (SPOT) method uses Brenier's theorem [6] to characterize the optimal transport map as the gradient of a convex potential function, i.e. $f = \nabla \phi$ [7]. The potential $\phi$ is parameterized by a sum of basis functions $\rho$:

$$
\phi(x) = \frac{1}{2}|x|^2 - \sum_{p \in Z} c(p) \rho(x - p)
$$

where $c$ are the coefficients to be determined via optimization, and $Z$ is the pixel grid for the image. For simplicity, the basis function $\rho$ is a radially symmetric Gaussian of standard deviation $\sigma$, where $\sigma$ is user-defined.

Rather than solving the Monge objective function directly, the solution is approximated by iteratively optimizing the following:

$$
\psi(c) = \frac{1}{2} \int_X \big( \det(D(x))I_1(f(x))- I_0(x) \big)^2 dx
$$

## Load Sample Data

In order to compare the different optimal transport methods, we must first load some images from which to compute the transport maps. The data we are going to use in this tutorial are from the OASIS cross-sectional brain MRI database. Due to usage restrictions, these images cannot be distributed with the *optimaltransport* package, but any images could be used for this example. The reference image $I_0$ used in this tutorial is the mean of the entire brain image dataset.

```python
from optrans.datasets import oasis

# Load the reference image
img0 = oasis.load_img0(fpath='../optrans/datasets/oasis_img0.npy')

# Load the OASIS data and select a single image to use
data, _, _ = oasis.load_data(fpath='../optrans/datasets/oasis_data.npz')
img1 = data[0]
```

In optimal transport, the signals (images) $I_0$ and $I_1$ must be strictly-positive probability density functions (PDF). Therefore, we have to convert the images into PDFs that integrate (or sum) to the same value. Fortunately, we can use the `signal_to_pdf()` function in the *optimaltransport* package to do this for us.

```python
from optrans.utils import signal_to_pdf

# Convert images to PDFs
img0 = signal_to_pdf(img0, sigma=1., total=100.)
img1 = signal_to_pdf(img1, sigma=1., total=100.)
```

In this example, we force the sum of the pixel intensities to be 100, and we smoothed the images slightly using a Gaussian filter with a standard deviation of 1 pixel.

## Initializing the Methods

To simplify the plotting procedure later on, we're going to initialize the optimal transport methods in a list.

```python
from optrans.continuous import RadonCDT, CLOT, VOT2D, MultiVOT2D, SPOT2D

# Initialise optimal transport methods
methods = [RadonCDT(theta=np.arange(180)),
           CLOT(lr=1e-7, tol=1e-4),
           VOT2D(alpha=0.001),
           MultiVOT2D(n_scales=3, alpha=0.001),
           SPOT2D(sigma=2.)]
```

Although we have explicitly stated some of the hyper-parameters, the `RadonCDT` and `SPOT2D` methods typically require very little parameter tuning. In fact, we're actually using the default parameter values for those two methods. We have specified a smaller `alpha` parameter than the default (0.01) for the `VOT2D` and `MultiVOT2D` algorithms, in order to help guide them towards a better match between $I_0$ and $I_1$. A lower `alpha` will reduce the penalty on the curl, and as a result, the transport map may not be truly optimal. From experience, the `CLOT` method is very difficult to initialize correctly, so good luck using that one...

## Computing Transforms

Once the optimal transport methods have been initialized we can compute the LOT of $I_1$ with respect to the reference $I_0$. Since the methods all share the `transport_map_` attribute, we can use the mapping to to reconstruct our original inputs:

* `img0_recon`: $I_0 = \det(D_f) I_1 \circ f$.
* `img1_recon`: $I_1 = \det(D_f)^{-1} I_0 \circ f^{-1}$.

We will also store the `displacements_` attribute so that we can plot it later.

```python
# Initialize recontructions and displacements
img0_recon = []
img1_recon = []
displacements = []

# Compute LOT transform and reconstruct images using transport map
for method in methods:
    lot = method.forward(img0, img1)
    img0_recon.append(method.apply_forward_map(method.transport_map_, img1))
    img1_recon.append(method.apply_inverse_map(method.transport_map_, img0))
    displacements.append(method.displacements_)
```

## Plotting Results

Now that we have computed the transport maps, we can plot the reconstructed images. We will also make use of the `plot_displacements2d()` function in the *optimaltransport* package in order to visualize the displacements computed by the different methods. It should be noted that, due to the Radon transform, the `RadonCDT` does not produces visually interpretable displacements. Therefore, we'll just ignore them!

```python
from optrans.visualization import plot_displacements2d

# Intensity ranges for plotting
vmin = min(img0.min(), img1.min())
vmax = max(img0.max(), img1.max())

# Initialise figure and plot original images
fig, ax = plt.subplots(len(methods)+1, 3, sharex=True, sharey=True, figsize=(9,20))
ax[0,0].imshow(img0, vmin=vmin, vmax=vmax)
ax[0,0].set_frame_on(False)
ax[0,0].set_title("$I_0$")
ax[0,0].set_ylabel("Original")
ax[0,1].imshow(img1, vmin=vmin, vmax=vmax)
ax[0,1].set_frame_on(False)
ax[0,1].set_title("$I_1$")
ax[0,2].axis('off')

for a,method,i0r,i1r,disp, in zip(ax[1:], methods, img0_recon, img1_recon, displacements):
    name = method.__class__.__name__
       
    # Plot images and displacements
    if name == 'RadonCDT':
        a[0].imshow(i0r)
        a[1].imshow(i1r)
    else:
        a[0].imshow(i0r, vmin=vmin, vmax=vmax)
        a[1].imshow(i1r, vmin=vmin, vmax=vmax)
        plot_displacements2d(disp, ax=a[2], count=20)

    # Set label
    a[0].set_ylabel(name)

    # Remove frame from plots
    a[0].set_frame_on(False)
    a[1].set_frame_on(False)
    a[2].set_frame_on(False)

# Display figure
plt.show()
```
